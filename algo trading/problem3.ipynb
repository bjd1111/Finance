# qp solver funtion

```python
from cvxopt import matrix, solvers,blas


def optimal_port(expected_return):
    
    solvers.options['show_progress'] = False     #disable progress
    

    symbol = list(C.index)
    C_s = C.as_matrix()             #convert to matrix
    
    data = stock_data[stock_data['Symbol'].isin(symbol)]      #select stock data based on symbol
    #print data
    ret_T = np.matrix(data['EstReturn %'])                     #get returns 
    #ret_T
    
    n = len(C_s)
    
    # w'Cw minimization using qp
    
    Q = matrix(C_s)                                         # covariance matrix
    p = matrix(np.full((n, 1), 0.))                         # return matrix, 0 here because returns are specfied outside,here only w'Cw
    G = matrix(np.r_[np.eye(n),-np.eye(n)])                 # inequality equation leftside
    h = matrix(np.r_[np.full((n, 1), 0.25),np.full((n, 1), -0.005)])  # inequality equation rightside
    A = matrix(np.r_[np.full((1, n), 1.0),ret_T] ,(2,n))              # equality equation leftside
    b = matrix([1.0,expected_return],(2,1))                           # equality equation rightside

    sol=solvers.qp(Q, p, G, h, A, b)
    #print sol
    std =  np.sqrt(np.dot(np.dot(np.array(sol['x']).T,C_s),np.array(sol['x'])))  # std = sqrt(w'Cw)
    returns = np.sum(ret_T*np.array(sol['x']))        # returns = sumproduct(w,R)
    #print returns
    #print std.shape
    return np.r_[np.array(sol['x']),std,np.array([[returns]])]                  # return w,std,return which is [n+2,1] matrix, n is the len(C)

#expected_return = 0.14
#print expected_return
#optimal_port(expected_return)
```

# protfolio with different expected return

```python

weights = np.zeros([len(C)+2,1])            #initial an empty array
expected_return = [0.165,0.16,0.155,0.15,0.145,0.14,0.135,0.13,0.125,0.12,0.115,0.11,0.105,0.10,0.09,0.08]


for j in expected_return:
    weights = np.c_[weights,optimal_port(j)]
    
weights = np.delete(weights, 0, 1)
weights = pd.DataFrame(weights.T ,columns=['w'+ str(i) for i in xrange(1,len(C)+1)]+['std','ret'],index=['portfolio'+ str(i) for i in xrange(1,len(expected_return)+1)])
weights
```

# For a  $100 Million Dollar Portfolio

### input data of size, volatility, POV

```python
capital = 100*10**6
w = np.array([weights.loc['portfolio14'][0:25]]).T
price = np.array([stock_data[stock_data['Group']==4]['Price']]).T
shares = w*capital / price

ADV = np.array([stock_data[stock_data['Group']==4]['ADV']],dtype='f8').T
size  = shares / ADV                                                     # size(%ADV) 
time = 1                                                                 # trade time % of day. relation to POV = size*(1-POV)/POV
POV = shares/(shares+time*ADV)                                           #shares/VOL = shares / (shares+time*ADV)
volatility = np.array([stock_data[stock_data['Group']==4]['Volatility']],dtype='f8').T

x = np.array(np.c_[size, volatility, POV])
```

### the market impact function from pb2 

```python
coeffs = [706.28378945135853, 0.34305784856765598, 0.56587621757055151, 0.75754853087902729, 0.97175923281985044]     #from problem 2
def MI_model(x):
    "market impact function"
    
    s = np.array(x[0])          #size
    v = np.array(x[1])          #volatility
    P = np.array(x[2])          #POV

    return coeffs[0]*(s**coeffs[1])*(v **coeffs[2]) * (coeffs[4]*(P **coeffs[3])+ (1-coeffs[4]))

MI_model(x[1])
```

### produce return and net return table including MI data

```python
mi = []
for i in xrange(len(x)):
    mi.append(MI_model(x[i]))
    
stocks = stock_data[stock_data['Group']==4] 
stocks.loc[:,'market_impact(bp)'] = mi
stocks.loc[:,'market_impact(decimal)'] = np.array(mi)/10000
stocks.loc[:,'market_impact(dollar)'] = np.array([mi]).T/10000*w*capital


stocks.loc[:,'invest'] = w * capital
stocks.loc[:,'return'] = stocks.loc[:,'invest'] * stocks.loc[:,'EstReturn %']
stocks.loc[:,'net_return'] = stocks.loc[:,'return']-stocks.loc[:,'market_impact(dollar)']
stocks
```

### expected return of portfolio

```python
#retrun
e_return_a1 = np.sum(stocks.loc[:,'return'])/np.sum(stocks.loc[:,'invest'])
e_return_a1
```

### expected Net Return of portfolio

```python
# expected Net Return %
e_n_return_a1  = np.sum(stocks.loc[:,'net_return'])/np.sum(stocks.loc[:,'invest'])
e_n_return_a1
```

### net change

```python
#change of return (because the expected return of portfolio is 10%, and now the net return of portfolio is e_n_return)
e_n_return_a1-e_return_a1

```

### market impact function for MI calculation

```python
def mi(weights,capital):
    w = weights
    price = np.array([stock_data[stock_data['Group']==4]['Price']]).T
    shares = w*capital / price

    ADV = np.array([stock_data[stock_data['Group']==4]['ADV']],dtype='f8').T
    size  = shares / ADV                                                     # size(%ADV) 
    time = 1                                                                 # trade time % of day. relation to POV = size*(1-POV)/POV
    POV = shares/(shares+time*ADV)                                           #shares/VOL = shares / (shares+time*ADV)
    volatility = np.array([stock_data[stock_data['Group']==4]['Volatility']],dtype='f8').T

    x = np.array(np.c_[size, volatility, POV])
    market_impact = []
    for i in xrange(len(x)):
        market_impact = np.append(market_impact,MI_model(x[i,:]))
    #print market_impact  
    market_impact = market_impact.reshape(len(market_impact),1)
    
    return np.array(market_impact)


```

# the function for maxmizing net return

```python
def func(w):
    w = np.reshape(w,(len(w),1))
    return np.sum(mi(w,capital)/10000*w*capital-returns*w*capital)
```

```python
# the net return 
func(w)
```

```python

cons = ({'type': 'eq',
          'fun' : lambda x: np.array([np.sum(x) -1])},
        
       
        {'type': 'eq',
          'fun' : lambda x: np.array([np.sqrt(float(np.dot(np.dot(x.T,C),x))) - std])}
       )

bon = tuple((0.005, 0.25) for i in range(25))
```

```python
from scipy.optimize import minimize


res = minimize(func, w.tolist(), constraints=cons, method='SLSQP',bounds=bon, options={'disp': True})
print(res.x)
```

```python
new_w = np.reshape(res.x,(len(res.x),1))
new_w
```
