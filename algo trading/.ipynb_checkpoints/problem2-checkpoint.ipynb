```python
import pandas as pd
import numpy as np

MI = pd.read_csv('MI.csv',header=2)
MI
```

```python
MI.shape
```

```python
r = np.random.RandomState(1234)
```

```python
MI_s = MI.sample(n=10,replace =True,random_state=r)
MI_s
```

```python
import scipy.optimize as optimization
from scipy.optimize import leastsq

def solve(sample):
    Y = sample['Cost_bp'].values
    size = sample['Size'].values
    volatility = sample['Volatility'].values
    POV = sample['POV'].values
    LnMktCap = sample['LnMktCap'].values


    def model(size,volatility,POV, LnMktCap, coeffs):
        return coeffs[0]*(size**coeffs[1])*(volatility**coeffs[2])*(LnMktCap**coeffs[4]) *  (coeffs[5]*POV **coeffs[3]   + (1-coeffs[3] ))

        #coeffs[0] is a1
        #coeffs[1] is a2
        #coeffs[2] is a3
        #coeffs[3] is a4
        #coeffs[4] is a5
        #coeffs[5] is b1

    x0 = np.array([0, 0, 0, 0,0,0], dtype=float)  # An approximative initial solution that we can find from looking at the graph is for instance:

    def residuals(coeffs,size,volatility,POV, LnMktCap ,y):
        return y - model(size,volatility,POV, LnMktCap, coeffs)

    coef,sucess = leastsq(residuals, x0, args=(size,volatility,POV, LnMktCap,Y))
    return coef


print solve(MI_s)
```

```python
df = pd.DataFrame(columns=['a1','a2','a3','a4','a5','b1'])
for i in xrange(10):
    MI_s = MI.sample(n=10,replace =True)
    result = pd.DataFrame(np.reshape(solve(MI_s),(1,6)),columns=['a1','a2','a3','a4','a5','b1'])
    df = df.append(result)
df = df.reset_index(drop=True)
df
```
