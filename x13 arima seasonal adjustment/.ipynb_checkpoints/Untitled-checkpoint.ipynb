```python

import matplotlib  
#matplotlib.use('Agg')  
#%matplotlib inline

from matplotlib.pyplot import plot,savefig  
from urllib2 import urlopen
import pandas as pd
import os
import statsmodels.api as sm
import numpy as np

PATH =r'C:\Users\bjd\Desktop\x13as'
os.chdir(PATH)




path = "ftp://ftp.cpc.ncep.noaa.gov/htdocs/degree_days/weighted/daily_data/"


start = 1990
end = 2000


file = path + str(start) + "/Population.Heating.txt"
u = urlopen(file)


def data_process(f):
    data=str(f.read()).split("'")[0]
    
    b = data.split("\n")
    time = b[3].split("|")[1:]
    temp = b[4:]
    temp=[x.split("|") for x in temp ]
    time=pd.DataFrame(time,columns=["time"])
    time["time"]=pd.to_datetime(time["time"])
    temp=pd.DataFrame([x[1:] for x in temp ],index=[x[0] for x in temp]).transpose()

    temp = temp.apply(pd.to_numeric)
    #print temp
    
    average_temp = temp["CONUS"]

    s = pd.Series(average_temp.values, index=time['time'])


    return s



final_data=data_process(u)

for year in range(start+1,end+1):
    n = path + str(year) + "/Population.Heating.txt"
    data_n = urlopen(n)

    final_data = pd.concat([final_data,data_process(data_n)])



s = final_data
s = s.resample('M', convention='start').mean()

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#plt.axis([1990,2000,8,15])

res = sm.tsa.seasonal_decompose(s)
plt.plot(res.trend)
  
```
