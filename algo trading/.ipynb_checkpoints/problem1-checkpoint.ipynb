# import data

```python
import pandas as pd
import numpy as np
price = pd.read_csv('price.csv')
ff3 = pd.read_csv('ff3.csv',header=2)
stock = pd.read_csv('stock_data.csv',header=2)
```

# new data return added

```python
price['return'] =  np.log(price['Close']) -np.log(price['Close'].shift(-1))
price.ix [price.Symbol!= price.Symbol.shift(-1),'return'] = np.nan 
#price[price['Symbol']=='CAT']
```

```python
price.shape
248976/504.
```

# ols regression function 

```python
from pandas.stats.api import ols
i=0
result = pd.DataFrame(columns=['Symbol','b0','b1','b2','b3','SeY','MSE'])
ff3 = ff3[0:503]
for n in xrange(494):
    
    df = pd.DataFrame(price['return'][i:i+504].values,columns= ['Rt'],index=xrange(504))
    RT = df['Rt']-ff3['RF']
    RT = RT[0:503]
    
    res = ols(y=RT, x=ff3[['Mkt-RF','SMB','HML']])
    SeY = res.rmse
    b1,b2,b3,b0 = res.beta
    #print res.summary
    MSE = SeY**2
    data = pd.DataFrame([[price['Symbol'][i], b0,b1,b2,b3,SeY,MSE]],columns=['Symbol','b0','b1','b2','b3','SeY','MSE'])
    result = result.append(data)
    i = i + 504
result = result.reset_index(drop=True)
result
```

# stock for our group

```python
#
symbol = list(stock[stock.Group==4]['Symbol'])
symbol
```

```python

data = pd.DataFrame(result[result['Symbol'].isin(symbol)])    #select data
data = data.reset_index(drop=True)                #reindex data
data.set_index('Symbol',inplace=True)           #transpos data 
data = data.T
data
```

# calculate covariance matrix

```python
B = np.matrix(data[1:4])
B_prime =  B.T
CovF = np.matrix(ff3.ix[:,2:5].cov())

idioC = np.matrix(np.diag(list(data.loc['MSE'])))

C_day = B_prime*CovF*B +idioC
C_annual = C_day*250
C_day
```

```python
C_annual
```

# output for further use

```python

a = pd.DataFrame(C_annual,columns=symbol,index=symbol)
a.to_csv('c.csv')
```
