```python
# Exercise: Compute the Black Scholes Option price when the intial price is 10, interest
# rate is 5%, volatility=0.25 and T=2 years. Use 10000 random draws 
rn = npr.standard_normal(10000)
z = rn
r = 0.05
sigma = 0.25
T = 2
S0 = 10
St = S0*np.exp((r-0.5*sigma**2)*T+sigma*np.sqrt(T)*z)

# Plot the distribution of the simulation, which distribution does it look like? Can you
# simulate the Black Scholes pricing directly using that distribution? 

plt.hist(St,bins=50)
plt.xlabel('index level')
plt.ylabel('frequency')
plt.grid(True)
# it looks like log normal

```

```python
# directly using log normal distribution

St1 = S0*npr.lognormal((r-0.5*sigma**2)*T,sigma*np.sqrt(T),size = 10000)

plt.hist(St1,bins=50)
plt.xlabel('index level')
plt.ylabel('frequency')
plt.grid(True)
```
