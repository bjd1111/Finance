```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

coeffs = [698.74940594355189, 0.3429159597976898, 0.55893058943625173, 0.74906715225239062, 0.97462393173698603]     #from problem 2
def MI_model(x):
    "market impact function"
    
    s = np.array(x[0])          #size
    v = np.array(x[1])          #volatility
    P = np.array(x[2])          #POV

    return coeffs[0]*(s**coeffs[1])*(v **coeffs[2]) * (coeffs[4]*(P **coeffs[3])+ (1-coeffs[4]))


size   = np.reshape(np.linspace(0,10,100),(100,1))
volatility = np.full((100,1),4)
POV = np.full((100,1),5)
x = np.array(np.c_[size, volatility, POV])
a = []
for i in xrange(100):
    a.append(MI_model(x[i]))
plt.plot(a)

```
