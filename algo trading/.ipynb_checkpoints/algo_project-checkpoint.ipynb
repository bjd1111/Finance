```python
import pandas as pd
import numpy as np
price = pd.read_csv('price.csv')
ff3 = pd.read_csv('ff3.csv',header=2)
```

```python
price['return'] = np.log(price['Close']) - np.log(price['Close'].shift(1))
price.ix [price.Symbol!= price.Symbol.shift(1),'return'] = np.nan  
```

```python
from sklearn.linear_model import LinearRegression
```

```python
price.shape
248976/504.
```

```python
i = 0
for n in xrange(494):
    
    df = pd.DataFrame(price['return'][i:i+504].values,columns= ['Rt'],index=xrange(504))
    RT = df['Rt']-ff3['RF']
    RT = RT.fillna(0)
    lr = LinearRegression()
    lr.fit(ff3[['Mkt-RF','SMB','HML']],y=RT) 
    print n,price['Symbol'][i] , 'coef', lr.coef_ , 'intercept', lr.intercept_
    i = i + 504
```
