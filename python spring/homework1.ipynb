```python
#get data from web

import urllib
import datetime
import numpy as np
import pandas as pd

### get the url for the downloadable csv-file data
def make_url(ticker_symbol, start_date, end_date):
    base_url = "http://ichart.finance.yahoo.com/table.csv?s="
    # print ticker_symbol
    a = start_date
    b = end_date
    dt_url = '%s&a=%d&b=%d&c=%d&d=%d&e=%d&f=%d&g=d&ignore=.csv'% (ticker_symbol, a.month-1, a.day, a.year, b.month-1, b.day,b.year)
    return base_url + dt_url

### create a path to save the file downloaded from web
#mac
#output_path = "/Users/apple/Downloads/lecture4" ## <======Please change to your own directory
#win
output_path = "C:/Users/bjd/Desktop/data" ## <======Please change to your own directory


def make_filename(ticker_symbol):
    return output_path + "/" + ticker_symbol + ".csv"
    

### pull the data out from web and store the file in the path that designed above
def pull_historical_data(ticker_symbol, start_date, end_date):
    try:
        urllib.urlretrieve(make_url(ticker_symbol, start_date, end_date), make_filename(ticker_symbol))
    except urllib.ContentTooShortError as e:
        outfile = open(make_filename(ticker_symbol), "w")
        outfile.write(e.content)
        outfile.close()




```

```python
data = dict()
for symbol in symbols:
    # create a new array in this slot
    data[symbol] = pd.read_csv('%s.csv' %symbol, sep=',', na_values=".")

#print data['AAPL']

#market data
pull_historical_data('SPY',start,end) ###<=== get historical time-series data for  HPQ
np.genfromtxt('SPY.csv', delimiter=",", skip_header=1) 
SPY = pd.read_csv('SPY.csv', sep=',', na_values=".")
SPY
```

```python
#Converting to returns
# assume rf = 0, in practice, rate of return on bills is on the order of only about .01% per day,  so total and excess returns are almost indistinguishable. 

def returns(data):
        data['excess_return'] = np.log(data['Adj Close']/data['Adj Close'].shift(1))-rf

returns(SPY)

for i in data:
    returns(data[i])

data['AAL']
```

```python
import statsmodels.api as sm

def reg(y,X):
    X = sm.add_constant(X)
    res = sm.OLS(y, X, missing='drop').fit()

    #print(res.summary())
    #print np.std(res.resid)
    alpha,beta = res.params
    return (alpha,beta, np.std(res.resid))
            
print reg(data['AAPL']['excess_return'],SPY['excess_return'])
a,b,c = reg(data['AAPL']['excess_return'],SPY['excess_return'])
```

# loop all stocks

```python
result = np.empty( shape=(0, 3))
for i in data:
    reg(data[i]['excess_return'],SPY['excess_return'])
    #print i
    #print reg(data[i]['excess_return'],SPY['excess_return'])
    result = np.r_[result,[reg(data[i]['excess_return'],SPY['excess_return'])]]

data1 = pd.DataFrame(result,columns=['alpha','beta','sigma(firm-specific)'],index=data.keys())
data1
```

```python
data2 = pd.DataFrame(data1['alpha']/data1['sigma(firm-specific)']**2,columns=['alpha/var'])
data2['weights'] = data2['alpha/var']/np.sum(data2['alpha/var'])
data2.loc['Total'] = data2.sum(numeric_only=True)
data2
```

```python
# Hence the alpha and beta of the active portfolio are 
alpha_p = np.dot(data2['weights'][:-1],data1['alpha'])
beta_p = np.dot(data2['weights'][:-1],data1['beta'])
#Residual variances are
var_residual_p =  np.dot(data2['weights'][:-1]**2,data1['sigma(firm-specific)']**2)
std_residual_p = np.sqrt(var_residual_p)
std_residual_p
```

```python
#adjusted weight for beta
w0 = (alpha_p/var_residual_p)/((ErM -rf)/(SigmaRM**2))

w_star = w0/(1+(1-beta_p)*w0)
w_star
```

```python
#Position in the index portfolio is 
weight_index = 1- (w_star)
weight_index
```

```python
# To calculate the Sharpe ratio for the optimal risky portfolio, 
# we compute the information ratio for the active portfolio and Sharpe’s measure for the market portfolio. 
# The information ratio for the active portfolio is computed as follows 

A = alpha_p/std_residual_p
A**2
```

```python
# Hence, the square of the Sharpe ratio (S) of the optimized risky portfolio is: 
S_sq = ( (ErM-rf)/SigmaRM  )**2 + A**2
S = np.sqrt(S_sq)
S
```

```python
market_S = (ErM-rf)/SigmaRM  
market_S
```

```python
# To calculate the makeup of the complete portfolio, first compute the beta, the mean excess return, 
# and the variance of the optimal risky portfolio: 

beta_cp = weight_index + ( w_star*beta_p)
mean_excess_return = w_star*alpha_p+beta_cp*ErM
var_cp = beta_cp**2*SigmaRM**2+w_star**2*var_residual_p
std_cp = np.sqrt(var_cp)
std_cp
```

```python
# with a coefficient of risk aversion of 2.8?     
# the optimal position in this portfolio is
aversion = 2.8
y = mean_excess_return/(0.01*aversion*var_cp*1000)
y
```
