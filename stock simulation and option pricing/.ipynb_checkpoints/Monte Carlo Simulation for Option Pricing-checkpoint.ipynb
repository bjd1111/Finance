```python
# -*- coding: utf-8 -*-

from math import exp 
from random import gauss
from matplotlib import pyplot

# Graphic set up
pyplot.xlabel('Time, steps at ' + str(steps))
pyplot.ylabel('Price, strike at $%.2f' % strike)
pyplot.axhline(strike, linewidth = 4, color = 'r', alpha = 0.2)
pyplot.axvline(steps, linewidth = 4, color = 'b', alpha = 0.2)
pyplot.grid(True)

# Process parameters

K = 60
steps = 252

S0 = 60
mu = .05
sigma = 0.2
T = 1.0
attempts = 1000

profit = []
# Monte Carlo simulation
for trials in range(attempts):
    prices = [S0]
    for day in range(steps):
        prices.append(prices[day] * exp((mu - .5 * sigma ** 2) * (T / steps) + 
        sigma * (T/ steps) ** .5 * gauss(0, 1)))
    
    pyplot.plot(prices)
    profit.append(max(0, prices[steps] - K))


# Calculate average profitability

average = reduce(lambda x, y : x + y, profit) / attempts

pyplot.title('Option price $%.2f ' % average)


pyplot.show()
```
